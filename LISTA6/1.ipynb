## Algorytm KMeans będzie miał problemy z danymi, w których różne featury mają różne skale np. wiek i zarobki. Feature z większą skalą dostaje większą "wagę". Przy różnicy wielkości kilku rzędów algorytm będzie grupował prawie tylko po featurze o największej skali. Standaryzacja naprawia ten problem, ponieważ sprowadza wszystkie featury do tej samej skali.
